```python
dft = df.copy()
```

```python
len(dft['Used name'])
```

```python
len(dft[dft['Used name'].str.contains('•')])
```

```python
dft['Name'] = dft['Used name'].str.replace("•", " ")
```

```python
dft.head()
```

```python
dft[['Height(CM)','Weight(KG)']] = dft['Measurements'].str.split('/', expand=True)
```

```python
dft[['Height(CM)','Weight(KG)']].info()
```

```python
#len(dft['Height(CM)'].str.contains(' cm'))
#len(dft['Weight(KG)'].str.contains(' kg'))
#dft['Height(CM)'] = dft['Height(CM)'].str.replace(' cm','')
#dft['Weight(CM)'] = dft['Weight(KG)'].str.replace(' kg','')
#dft['Height(CM)'] = pd.to_numeric(dft['Height(CM)'])
#ValueError: Unable to parse string "120 kg" at position 165
#dft.loc[dft['Height(CM)'] == '120 kg']
```

Many columns where we only have CM or KG

```python
dft[~dft['Measurements'].str.contains('/',na=False) & dft['Measurements'].notna()]
```

```python
selected = dft['Height(CM)'].str.contains('kg', na=False)
dft.loc[selected, 'Weight(KG)'] = dft.loc[selected, 'Height(CM)']
dft.loc[selected, 'Height(CM)'] = np.nan
dft['Height(CM)'] = pd.to_numeric(dft['Height(CM)'].str.strip(' cm'), errors='coerce')
dft['Weight(KG)'] = pd.to_numeric(dft['Weight(KG)'].str.strip(' kg'), errors='coerce')
```

```python
dft.head()
```

```python
date_pattern = r'(\d+ \w+ \d{4}|\d{4}|\w+ \d{4})'

dft['born_date'] = dft['Born'].str.extract(date_pattern)
dft['born_date'] = pd.to_datetime(dft['born_date'], format='mixed', errors='coerce')

dft['died_date'] = dft['Died'].str.extract(date_pattern)
dft['died_date'] = pd.to_datetime(dft['died_date'], format='mixed', errors='coerce')

dft.head()
```

```python
dft[~dft['Born'].str.match(date_pattern, na=False)&dft['Born'].notna()]
```

```python
dft.info()
```

```python
dft['Born'].head()
```

```python
#location_pattern = r'in ([\w\s+]),\s+([\w\s+])\+\((\w+)\)'
location_pattern = r'in ([\w\s-]+), ([\w\s-]+) \((\w+)\)'
dft[['born_city','born_region','born_country']] = dft['Born'].str.extract(location_pattern, expand=True)
```

```python
dft.head()
```

```python
dft.columns
```

```python
columns_to_keep = ['athlete_id', 'Name', 'born_date', 'born_city', 'born_region', 'born_country', 'NOC', 'Height(CM)', 'Weight(KG)', 'died_date']
df_1 = dft[columns_to_keep]
```

```python
df_1.head()
```

Part Two

```python
cols = ['ID', 'FN', 'Date', 'City', 'Region', 'Country', 'NOC', 'Height', 'Weight', 'D_O_D']
```

```python
df_1.columns = cols
```

```python
df_1.head()
```

```python
df_1['fN'] = df_1['FN'].str.split(' ').str[0]
```

```python
df_1['lN'] = df_1['FN'].str.split(' ').str[1]
```

```python
df_1.head()
```

```python
df_1.drop(columns='FN', inplace=True)
df_1 = df_1[['ID', 'fN', 'lN', 'Date', 'City', 'Region', 'Country', 'NOC', 'Height', 'Weight', 'D_O_D']]
df_1.head()
```

```python
df_1.info()
```

```python
df_1['Y'] = df_1['Date'].dt.year
df_1['M'] = df_1['Date'].dt.month
df_1['D'] = df_1['Date'].dt.day
```

```python
df_1.head()
```

```python
df_1.drop(columns='Date',inplace=True)
df_1 = df_1[['ID','fN','lN','Y','M','D','City','Region','Country','NOC','Height','Weight','D_O_D']]
```

```python
df_1.head()
```

```python
df_1['YD'] = df_1['D_O_D'].dt.year
df_1['MD'] = df_1['D_O_D'].dt.month
df_1['DD'] = df_1['D_O_D'].dt.day
```

```python
df_1.head()
```

```python
df_1.drop(columns='D_O_D',inplace=True)
```

```python
df_1.head()
```

```python
nocs = pd.read_csv('noc_regions.csv')
```

```python
df_1 = pd.merge(df_1,nocs,left_on='Country',right_on='NOC', how='left')
```

```python
df_1.head()
```

```python
df_1 = df_1.rename(columns={'region':'country'})
```

```python
df_1.drop(columns=['Country','NOC_x','NOC_y','notes'],inplace=True)
```

```python
df_1 = df_1.rename(columns={'country':'Country'})
```

```python
df_1 = df_1[['ID','fN','lN','Y','M','D','City','Region','Country','Height','Weight','YD','MD','DD']]
```

```python
df_1.head()
```

```python
df_1 = df_1.rename(columns={'Height':'H(CM)'})
df_1 = df_1.rename(columns={'Weight':'W(KG)'})
```
