```python
mean1 = df['Available Points'].mean()
mean2 = df['Available Points'].mean()
mean3 = df['Available Points'].mean()
mode1 = df['Available Points'].mode()
mode2 = df['Available Points'].mode()
mode3 = df['Available Points'].mode()
median1 = df['Available Points'].median()
median2 = df['Available Points'].median()
median3 = df['Available Points'].median()
print("Mean of Available Points: ", mean1)
print("Mean of Total Earned Points: ", mean2)
print("Mean of Points Spent: ", mean3)
print("")
print("Mode of Available Points: ", mode1)
print("Mode of Total Earned Points: ", mode2)
print("Mode of Points Spent: ", mode3)
print("")
print("Median of Available Points: ", median1)
print("Median of Total Earned Points: ", median2)
print("Median of Points Spent: ", median3)
```

```python
highest_available = max(df['Available Points'].sort_values(ascending=False))
print("Highest Available Points for User: ", highest_available)
```

```python
lowest_available = min(df['Available Points'].sort_values())
print("Lowest Available Points for User: ", lowest_available)
```

```python
best_user = df.loc[df['Available Points'] == highest_available]
best_user_id = best_user['User ID'].iloc[0]
best_user_state = best_user['State'].iloc[0]
best_user_city = best_user['City'].iloc[0]
print("Best User ID: ", best_user_id)
print("Best User State: ", best_user_state)
print("Best User City: ", best_user_city)
```
